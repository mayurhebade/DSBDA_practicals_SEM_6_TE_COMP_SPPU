```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
# Data Formatting and Normalization
scaler = MinMaxScaler()
```

```python
# Normalize numerical columns
num_columns = ['Year of Manufacture']
df[num_columns]=  scaler.fit_transform(df[num_columns])
```

```python
# Convert Categorical Variables to Quantitative Variables
encoder =LabelEncoder()
var_columns=['First Name','Last Name','Country','Car Brand','Car Model','Car Color','Credit Card Type']
for col in var_columns:
    df[col] = encoder.fit_transform(df[col])
```

```python
df.to_csv('new_processed_cars_all_numeric.csv', index=False)
```
