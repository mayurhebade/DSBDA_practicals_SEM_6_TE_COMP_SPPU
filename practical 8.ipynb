```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df['Age'].fillna(df['Age'].median())
df[['Cabin', 'Embarked']] = df[['Cabin', 'Embarked']].fillna('Nan')
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Fare'], bins=30, edgecolor='black')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Number of Passengers')
plt.ylabel('')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
fig, axis = plt.subplots(1,2)
sns.histplot(ax = axis[0], data = df, x='Age', multiple = 'dodge', shrink = 0.8, kde = True)
sns.histplot(ax = axis[1], data = df, x='Fare', multiple = 'dodge', shrink = 0.8, kde = True)
plt.show()
```
